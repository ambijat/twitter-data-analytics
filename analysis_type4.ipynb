# This page gives a new method of dataframe separation for `positive` and `negative` tweets based on `negative-words2.exd` file.

```python
import pandas as pd 
import glob
import os
```

#### Compiled tweets can also be procured from processed2/tweets2.csv

```python
process_path = input("Enter the path for raw data file: ")
all_files = []
all_files = glob.glob(process_path + "/datewise/" + "*.csv")
print(list(all_files))
```

```python
tweets = pd.concat((pd.read_csv(f, lineterminator='\n', sep = ';') for f in all_files))
tweets.head(10)
```

```python
tweets2 = tweets.iloc[:,[0, 1, 2, 3, -1]]
tweets2.head()
```

```python
len(tweets2)
```

```python
negative = []
n_file = "negative-words2.exd"
print(n_file)
```

```python
with open (n_file, "r") as nwfile:
    for line in nwfile:
        negative.append(line.strip())
```

```python
type(negative)
```

```python
len(negative)
```

```python
processed2_path = process_path + "/" + 'processed2'
print(processed2_path)
if not os.path.exists(processed2_path):
    os.makedirs(processed2_path)
```

```python
tweets3 = tweets2[tweets2['full_text'].str.contains(r'(?:{})'.format('|'.join(negative)), regex=True, na=False)].copy()
```

```python
len(tweets3)
```

```python
tweets3.head()
```

```python
tweets3['RT'] = tweets3['full_text'].str.contains(r"^(?:RT\s@\w+:)", case=False) == True
```

```python
tweets3.head(20)
```

```python
negative_tweets = tweets3
```

```python
negpath = os.path.join(processed2_path, 'negative_tweets4.csv')
print(negpath)
negative_tweets.to_csv(negpath, index = False, header = True, encoding = 'utf-8', sep = ';')
```

```python
tweets4 = tweets2[~tweets2['full_text'].str.contains(r'(?:{})'.format('|'.join(negative)), regex=True, na=False)].copy()
```

```python
tweets4.head(20)
```

```python
len(tweets4)
```

```python
positive_tweets = tweets4
```

```python
total_tweets = len(negative_tweets) +  len(positive_tweets)
print(total_tweets)
```

```python
pospath = os.path.join(processed2_path, 'positive_tweets4.csv')
print(pospath)
positive_tweets.to_csv(pospath, index = False, header = True, encoding = 'utf-8', sep = ';')
```
