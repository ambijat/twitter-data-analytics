```python
text_negative = pd.read_csv(p2_file, names = ['text'], header = None)
text_positive = pd.read_csv(p6_file, names = ['text'], header = None)
```

```python
text2 = pd.DataFrame(text_negative.text.unique())
```

```python
text2[:20]
```

```python
text2.rename(columns={0 : 'text'}, inplace=True)
```

```python
text2 = text2.reindex((text2.text.str.len()).argsort()).reset_index(drop=True)
```

```python
text2[:20]
```

```python
mask = (text2['text'].str.len() >= 270)

if len(mask == 0):
    mask = (text2['text'].str.len() >= 100)
```

```python
text3 = text2.loc[mask].reset_index(drop = True)
```

```python
text3[:20]
```

```python
len(text3)
```

```python
p3_file = path_to_process_file + "/" + "process_type7.tmp"
print(p3_file)
p4_file = path_to_process_file + "/" + "negative_tweets_cleaned.txt"
print(p4_file)
```

```python
text3['text'].to_csv(p3_file, header = False, index = False)
```
